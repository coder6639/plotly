```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),
]
```

```python
df = pd.DataFrame(prices, columns=['month', 'price'])
```

```python
df
```

```python
df.set_index('month')
```

```python
df['price in USD'] = [price / 4 for price in df['price']]
df
```

```python
df.set_index('month')
```

```python
plt.plot(df['month'], df['price in USD'], color='red', linestyle='dashed')
plt.xlabel('months')
plt.ylabel('USD')
plt.title('Price of goods')
```
